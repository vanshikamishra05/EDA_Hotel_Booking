## **To start with the very basic of data cleaning, let's find out if any of the columns have any Null or missing values and replace those with integers.**

```python
Hotel_Booking_df.isnull().sum()
```

```python
Hotel_Booking_df['Country'].fillna('xyz',inplace= True)
Hotel_Booking_df['Children'].fillna(0 , inplace = True)
Hotel_Booking_df['Children'] = Hotel_Booking_df['Children'].astype(int)
```

## **Removing columns which are not required for analysis and clubbing others.**

```python
#Checking correlation between columns to remove columns that are not necessary

sns.heatmap(Hotel_Booking_df[['Type of Hotel', 'Year', 'Month', 'Reservation Date', 'Arrival week',
       'Arrival Date', 'Weekend Stays', 'Weekday Stays', 'Meal',
       'Booking Payment', 'Adults', 'Children', 'Kids', 'Country',
       'Booking Type', 'Repeated Guest', 'Prev Cancel', 'History','Reserved Room', 'Assigned Room', 'Booking Changes', 'Waiting List',
       'Customer', 'Per Day Charges', 'Parking at premises',
       'Physical Challenged', 'Reservation Status', 'Cancellation',
       'Booking Done']].corr(), annot = True)
sns.set ({'figure.figsize':(28, 28)})
plt.title('Correlation Map for the Jobs Data', fontsize = 20)
plt.show()
```

we observe that columns are not highly correlated with each other so we can not delete rows on basis of that.

```python
# Creating a copy so as not to disturb original dataset.
HMbooking_df = Hotel_Booking_df[select_column].copy()
```

```python
HMbooking_df['Total_guests'] = HMbooking_df['Adults']+HMbooking_df['Kids']
```

```python
HMbooking_df = HMbooking_df.drop(['Adults', 'Kids'], axis=1)
```

```python
# removing rows which contain total guests as Zero.
HMbooking_df= HMbooking_df[HMbooking_df.Total_guests > 0]
```
